```python
from pathlib import Path #not sure what this library is. i guess maybe something for dealing with file paths?

all_txt_files = [] # Prepares to make a list of all the text files

for file in Path("txt").rglob("*.txt"): #not sure what rglob is doing
    all_txt_files.append(file.parent/file.name) #seems to be adding the path and file name of the text file

n_files = len(all_txt_files) # gets the number of items in the list it seems
print(n_files) #prints it out
```

```python
all_txt_files.sort() # this should sort the files in ascending order by month and date since i'm guessing that is how the files are named
print(all_txt_files[0]) #displays the first item in the list; idk why its a 'posix path' here; its not like that in the tutorial
all_txt_files[0] # the difference between print and not having print
```

```python
all_docs = [] #a new list for all the docs
#txt_file_as_string = "" # a new string; this was also not in the tutorial code so idk why i need it, or rather, why the tutorial code doesn't need it

for txt_file in all_txt_files: #iterates through the list of text files
    with open(txt_file) as f: # i can't quite explain whats going on here, but basically i think the with is a way to open the file and automatically close it when we're done. not sure it has any other uses. Then it just basically puts it into a variable called f. 
        print(txt_file) #prints out which file we are working on for debugging 
        txt_file_as_string = f.read() # reads each file as a string
    all_docs.append(txt_file_as_string) #then appends that string to the list of strings in all_docs

print(txt_file_as_string) #prints out that string
print(all_docs[0:3]) 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# not really sure what is going on here; i guess this is the setting for doing the analysis but like...
vectorizer = TfidfVectorizer(max_df=.65, min_df=1, stop_words=None, use_idf=True, norm=None)
transformed_documents = vectorizer.fit_transform(all_docs) #sparse matrix??
```

```python
print(transformed_documents) #O.O
```

```python
transformed_documents_as_array = transformed_documents.toarray() # "A numpy array is list-like but not exactly a list" takes the above and puts
# use this line of code to verify that the numpy array represents the same number of documents that we have in the file list
len(transformed_documents_as_array)
```

```python
import pandas as pd

# make the output folder if it doesn't already exist
Path("./tf_idf_output").mkdir(parents=True, exist_ok=True) #exist_ok=True i think probably means 'if it exists, it is ok to not complete the command'

# construct a list of output file paths using the previous list of text files the relative path for tf_idf_output
output_filenames = [str(txt_file).replace(".txt", ".csv").replace("txt/", "tf_idf_output/") for txt_file in all_txt_files]

# loop each item in transformed_documents_as_array, using enumerate to keep track of the current position
for counter, doc in enumerate(transformed_documents_as_array):
    # construct a dataframe
    tf_idf_tuples = list(zip(vectorizer.get_feature_names(), doc))
    one_doc_as_df = pd.DataFrame.from_records(tf_idf_tuples, columns=['term', 'score']).sort_values(by='score', ascending=False).reset_index(drop=True) # could figure it all out up to reset index bit

    # output to a csv using the enumerated value for the filename
    one_doc_as_df.to_csv(output_filenames[counter])
```
