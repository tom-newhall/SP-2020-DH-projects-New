```python
# first, determine the maximum legnth of an term in the dictionary; then we can begin our search for strings of that legnth
from bs4 import BeautifulSoup #idk what bs4 is, but beautiful soup is a parser for xml. we use this to, you know, parse the xml 

dfb_file = open('dingfubao.xml') #opens the file
dfb_soup = BeautifulSoup(dfb_file, 'xml') #BeautufilSoup parses the file; preparing it for use 
```

```python
for headword in dfb_soup("form"): #BeautifulSoup is real simple to find tags and stuff; this finds all the "form" tags which hold the headword of each entry
    headword = headword.text # saves the text of the 'form' tag in the variable headword
    print(headword, "\t", len(headword)) # prints the headword and its legnth in number of characters
```

```python
for entry in dfb_soup.find_all("entry"): #finds all the "entry" tags in the dictionary
    print("HW=", entry.form.text,"\t usg=", entry.sense.usg.text) #finds the text of the 'form' tag (i.e. the headword) within the 'entry' tag and the text of the 'usg' tag (i.e. the POS) within the 'sense' tagtag
```

```python
dict_with_usg = {} #makes a dictionary variable to store my data

for entry in dfb_soup.find_all("entry"): #finds all the "entry" tags in the dictionary
    headword = entry.form.text
    usg = entry.sense.usg.text
    dict_with_usg[headword] = usg
#    print("HW=", entry.form.text,"\t usg=", entry.sense.usg.text) #finds the text of the 'form' tag (i.e. the headword) within the 'entry' tag and the text of the 'usg' tag (i.e. the POS) within the 'sense' tagtag
print(dict_with_usg)
```
