```python
# ok well, i couldn't make it work with the xml parser way so we'll do this differently
# bit of code that finds the headwords from the dingfubao dictionary using regex
import re
file = open("../XSC-Analysis/dingfubao.xml", "r", encoding="utf-8")
data_from_file = file.read()
file.close()

file = open("dingfubao-headwords.txt", "w", encoding="utf-8")
for match in re.finditer(r"<form>(.+?)<\/form>", data_from_file):
    matched_text = match.group(1)
    print(matched_text)
    file.write(matched_text+"\n")
file.close()
```

```python
# puts the above DFB list of headwords back into a python list so you can use it

with open("dingfubao-headwords.txt", "r", encoding="utf-8") as dfb:
    dfb_list = dfb.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
dfb_list = [x.strip() for x in dfb_list]
print(len(dfb_list)) #prints the number of items in the list; the number of headwords
```
