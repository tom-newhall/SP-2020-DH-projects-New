# Very basic Classical Chinese tokenization

- Uses the Stanza Classical chinese package to do a basic tokenization, but the end result is that it really only breaks into characters; stopwords, pos tagging, punctuation getting rid of-ing, or any other fancy stuff will have to be implemented later, using a proper dictionary. Essentially, I really need to make my own tokenizer but getting this up and running was the main point. 

```python
#load the file to process
with open("T1804/T1804_ch9_line_breaks.txt", "r", encoding="utf-8") as ch9: # i guess the 'with'  is just a way to open the file a way that prevents errors from happening, and also closes it automatically, perhaps this is a better way to open files generally?
    ch9_text = ch9.readlines()  # reads each line as a string in a list

ch9_stripped = [x.strip() for x in ch9_text] #don't really understand this syntax, but it basically strips of the newlines and then puts the whole thing into a list; not sure if there is a way to do this without making it into a list first?
ch9_joined = ''.join(ch9_stripped) #this joins all the elements of the list into a long string. syntax really doesn't make sense to me, but the '' is just a way to say don't put any space between the things you are joining. It seems like it should be <string>.join('<what goes between the strings>') but w/e

#load the nlp program
import stanza
nlp = stanza.Pipeline('lzh', processors='tokenize,mwt,pos', tokenize_no_ssplit = True) #by doing 'no ssplit' it do it without breaking into sentences?. 

# process the file
doc = nlp(ch9_joined) 

# get just the tokens and print them out; there should only be one 'sentence' for the whole thing since we set tokenize_no_ssplit to True , so maybe there's a simpler way to do this that doesn't involve the first for loop
tokenized = []
for sentence in doc.sentences:
    for token in sentence.tokens:
        tokenized.append(token.text) #i guess this is one way to append to a list. Seems like there should be some kind of = expression in there but w/e

print(tokenized) #sorry not stopwords, pos tagging, punctuation getting rid of-ing, or any other fancy stuff. This is useless but I guess I figured  out how to do a useless thing. 
```

```python
first_sent = doc.sentences[0]
print(first_sent.words[1].xpos)
```

```python
for sentence in doc.sentences: #the whole doc is rendered as one sentence b/c the tokenize nossplilt tag and because its easier that way for now
    #print("Text is:", sentence.text)
    for word in sentence.words:
        print("Word is:", word.text, "\tPOS is:", word.upos) # so that's the right syntax finally
```
